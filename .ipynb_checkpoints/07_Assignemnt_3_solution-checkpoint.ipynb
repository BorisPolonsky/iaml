## Description of the dataset
In this assignment our goal is to recognize objects from 19 dierent visual classes (e.g. person, dog, cat, car, ...) in realistic scenes. The dataset consists of several thousands photographs harvested from the web. Each object of a relevant class has been manually annotated with a bounding box. Images can contain none, one or multiple objects of each class. We have prepared a [website](http://www.inf.ed.ac.uk/teaching/courses/iaml/2014/assts/asst3/images.html) where you can view the images.

Here we will focus on a single classiffication task: you will be required to classify images as to whether or not they contain a person. To save you time and to make the problem manageable with limited computational resources, we have preprocessed the dataset. We will use the [Bag of Visual Words](https://en.wikipedia.org/wiki/Bag-of-words_model_in_computer_vision) representation. That is, each image is represented by a 500 dimensional vector that contains the normalized count for each of 500 diffeerent visual words present in the respective image (a similar representation is used for the spambase dataset, just for real words). Visual words are based on [Scale-invariant feature transforms (SIFT)](https://en.wikipedia.org/wiki/Scale-invariant_feature_transform). SIFT features are essentially local orientation histograms and capture the properties of small image regions. They possess attractive invariance properties which make them well suited for our task (you can read more about SIFT features in [D.Lowe, IJCV 60(2):91- 110, 2004](http://link.springer.com/article/10.1023/B:VISI.0000029664.99615.94), but the details don't matter for the purpose of this assignment). Each SIFT feature is a 128 dimensional vector. From each image many SIFT features are extracted, typically > 2500 per image (features are extracted at regular intervals using a 15 pixel grid and at 4 different scales). To obtain visual words a representative subset of all extracted SIFT features from all images is chosen and clustered with k-means using 500 centres (such use of the k-means algorithm will be discussed in detail during the lecture). These 500 cluster centres form our visual words. The representation of a single image is obtained by first assigning each SIFT feature extracted from the image to the appropriate cluster (i.e. we determine the visual word corresponding to each feature by picking the closest cluster centre). We then count the number of features from that image assigned to each cluster (i.e. we determine how often each visual word is present in the image). This results in a 500 dimensional count vector for each image (one dimension for each visual word). The normalized version of this count vector gives the final representation of the image (normalized means that we divide the count vector by the total number of visual words in the image, i.e. the normalized counts sum to 1 for each image). Our dataset with all images is thus a $N \times 500$ dimensional matrix where `N` is the number of images. The resulting representation is much more compact and can be used directly to perform classiffication.

The full dataset has 520 attributes (dimensions). The first attribute (`imgID`) contains the image ID which allows you to associate a data point with an actual image. The next 500 attributes (`dim1`, ..., `dim500`) correspond to the normalized count vector. The last 19 attributes (`is_class`) indicate the presence of at least one object of a particular class in the image. In most of the experiments (unless explicitly noted otherwise) you will be asked to train classiffiers for classifying person vs. non-person images and only the `is_person` attribute and the 500 dimensional feature vector will be used. Do not use the additional class indicator attributes as features unless explicitly told to do so.

### Question 1.1
Load the datasets `train_images_partA.csv` and `valid_images_partA.csv` into two pandas DataFrame called `train_A` and `valid_A`. Display the number of data points and attributes in each of the two datasets.

```python
# Your code goes here
data_path_train = os.path.join(os.getcwd(), 'datasets', 'train_images_partA.csv')
data_path_valid = os.path.join(os.getcwd(), 'datasets', 'valid_images_partA.csv')
train_A = pd.read_csv(data_path_train, delimiter = ',')
valid_A = pd.read_csv(data_path_valid, delimiter = ',')
print('Training dataset, number of samples: ', train_A.shape[0], ' number of attributes: ', train_A.shape[1])
print('Validation dataset, number of samples: ', valid_A.shape[0], ' number of attributes: ', valid_A.shape[1])
```

```python
train_A.head(10)
```

```python
# Isolate required features
columns_of_interest = []
for dim in range(1,501):
    columns_of_interest.append('dim' + str(dim))
```

```python
# Get training/validation data
X_tr = train_A[columns_of_interest].as_matrix()
X_val = valid_A[columns_of_interest].as_matrix()
y_tr = train_A["is_person"].as_matrix()
y_val = valid_A["is_person"].as_matrix()
```

```python
# Histogram of target values in validation set
sns.distplot(y_val, kde=False, norm_hist=False)
print('Baseline accuracy score: ', accuracy_score(y_val, np.zeros_like(y_val)))
```

```python
# Off-the self LR
lr = LogisticRegression().fit(X_tr,y_tr)
print('Logistic regression accuracy score on validation set: ', lr.score(X_val, y_val))
```

```python
# Standardization
sc = StandardScaler().fit(X_tr)
X_tr = sc.transform(X_tr)
X_val = sc.transform(X_val)
```

```python
# Off-the self LR with standardization
lr = LogisticRegression().fit(X_tr,y_tr)
print('Logistic regression accuracy score on validation set (with standardizaation): ', lr.score(X_val, y_val))
```

```python
# Cross-validation
cv = KFold(n=train_A.shape[0], n_folds=3, shuffle=True, random_state=0)
```

```python
# LR: optimize regularization parameter C via CV
C_range = np.logspace(-5,5,20)
performance = np.zeros((len(cv), C_range.size))
for ii, (train, valid) in enumerate(cv):
    X_tr_tr = X_tr[train,:]
    X_tr_val = X_tr[valid,:]
    y_tr_tr = y_tr[train]
    y_tr_val = y_tr[valid]
    for jj, C in enumerate(C_range):
        clf = LogisticRegression(C=C).fit(X_tr_tr,y_tr_tr)
        performance[ii,jj] = clf.score(X_tr_val, y_tr_val)
```

```python
plt.figure()
plt.semilogx(C_range, performance.mean(axis=0))
plt.xlabel('Regularization parameter (C)')
plt.ylabel('Mean cross-validated classification accuracy')
C_opt = C_range[np.argmax(performance.mean(axis=0))]
print('Optimal regularization value (C): ', C_opt)
lr = LogisticRegression(C=C_opt).fit(X_tr,y_tr)
print('Accuracy score on validation set: ', lr.score(X_val, y_val))
```

```python
# Off-the shelf CV optimization for Logistic Regression
lrcv = LogisticRegressionCV().fit(X_tr,y_tr)
print('Accuracy score on validation set: ', lrcv.score(X_val, y_val))
print('Optimal regularization value (C): ', lrcv.C_[0])
```

```python
# Random forest classifier
rf = RandomForestClassifier(n_estimators=200).fit(X_tr, y_tr)
rf.score(X_val, y_val)
```

```python
# Feature importance in descending order
feat_order = np.argsort(rf.feature_importances_)[::-1] # In descending order
print(np.asarray(columns_of_interest)[feat_order][0:50])
```

```python
# SVC default settings: kernel comparisons
svc_lin = SVC(kernel='linear').fit(X_tr, y_tr)
svc_rbf = SVC(kernel='rbf').fit(X_tr, y_tr)
svc_poly = SVC(kernel='poly').fit(X_tr, y_tr)
print('Accuracy on validation set with SVM (linear kernel): ', svc_lin.score(X_val,y_val))
print('Accuracy on validation set with SVM (rbf kernel): ', svc_rbf.score(X_val,y_val))
print('Accuracy on validation set with SVM (3rd-order poly kernel): ', svc_poly.score(X_val,y_val))
```

```python
# Plot decision boundaries with different kernels
# TODO: write this as a function and ask them to input the classifiers and 2D input data
feats = [20, 75]
X_plot = X_tr[:,feats]

C = 1.0 # SVM regularization parameter
lin_svc = SVC(kernel='linear', C=C).fit(X_plot, y_tr)
rbf_svc = SVC(kernel='rbf', gamma='auto', C=C).fit(X_plot, y_tr)
poly_svc = SVC(kernel='poly', degree=3, C=C).fit(X_plot, y_tr)
clfs = [lin_svc, rbf_svc, poly_svc]

# create a mesh to plot in
h = .02 # step size in the mesh
x_min, x_max = X_plot[:, 0].min() - 1, X_plot[:, 0].max() + 1
y_min, y_max = X_plot[:, 1].min() - 1, X_plot[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
plt.figure(figsize=(15,5))
for i, clf in enumerate(clfs):
    plt.subplot(1, len(clfs), i + 1)
    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    # Plot also the training points
    plt.scatter(X_plot[:, 0], X_plot[:, 1], c=y_tr, cmap=plt.cm.Paired)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.xticks(())
    plt.yticks(())
plt.show()
```

```python
# SVC optimization C
C_range = np.logspace(-2, 3, 10)
score_svc_rbf_C = np.zeros((len(cv), C_range.size))
for ii, (train, valid) in enumerate(cv):
    for jj, C in enumerate(C_range):
        clf = SVC(kernel='rbf', C=C, gamma='auto').fit(X_tr[train,:], y_tr[train])
        score_svc_rbf_C[ii, jj] = clf.score(X_tr[valid,:], y_tr[valid])
```

```python
plt.figure()
plt.semilogx(C_range, score_svc_rbf_C.mean(axis=0))
plt.xlabel('Regularization parameter (C)')
plt.ylabel('Mean cross-validated classification accuracy')
C_opt = C_range[np.argmax(score_svc_rbf_C.mean(axis=0))]
print('Optimal regularization value (C): ', C_opt)
svc_rbf = SVC(kernel='rbf', C=C_opt, gamma='auto').fit(X_tr,y_tr)
print('Accuracy score on validation set: ', svc_rbf.score(X_val, y_val))
```

```python
# SVC optimization gamma
gamma_range = np.logspace(-5, 0, 10)
score_svc_rbf_gamma = np.zeros((len(cv), gamma_range.size))
for ii, (train, valid) in enumerate(cv):
    for jj, gamma in enumerate(gamma_range):
        clf = SVC(kernel='rbf', gamma=gamma).fit(X_tr[train,:], y_tr[train])
        score_svc_rbf_gamma[ii, jj] = clf.score(X_tr[valid,:], y_tr[valid])
```

```python
plt.figure()
plt.semilogx(gamma_range, score_svc_rbf_gamma.mean(axis=0))
plt.xlabel('Kernel coefficient parameter (gamma)')
plt.ylabel('Mean cross-validated classification accuracy')
gamma_opt = gamma_range[np.argmax(score_svc_rbf_gamma.mean(axis=0))]
print('Optimal kernel coefficient value (gamma): ', gamma_opt)
svc_rbf = SVC(kernel='rbf', gamma=gamma_opt).fit(X_tr,y_tr)
print('Accuracy score on validation set: ', svc_rbf.score(X_val, y_val))
```

```python
# SVC optimization C and gamma grid
C_2d_range = np.logspace(-2, 1, 4)
gamma_2d_range = np.logspace(-4, -1, 4)
score_svc_rbf_C_gamma = np.zeros((len(cv), C_2d_range.size, gamma_2d_range.size))
for ii, (train, valid) in enumerate(cv):
    for jj, C in enumerate(C_2d_range):
        for kk, gamma in enumerate(gamma_2d_range):
            clf = SVC(kernel='rbf', C=C, gamma=gamma).fit(X_tr[train,:], y_tr[train])
            score_svc_rbf_C_gamma[ii, jj, kk] = clf.score(X_tr[valid,:], y_tr[valid])
```

```python
plt.figure()
ax = sns.heatmap(score_svc_rbf_C_gamma.mean(axis=0), xticklabels=gamma_2d_range, yticklabels=C_2d_range, vmin=0.5, vmax=0.7)
plt.ylabel('Regularization parameter (C)')
plt.xlabel('Kernel coefficient parameter (gamma)')
idx_max = np.unravel_index(score_svc_rbf_C_gamma.mean(axis=0).argmax(), score_svc_rbf_C_gamma.mean(axis=0).shape)
C_opt = C_2d_range[idx_max[0]]
print('Optimal regularization value (C): ', C_opt)
gamma_opt = gamma_2d_range[idx_max[1]]
print('Optimal kernel coefficient value (gamma): ', gamma_opt)
svc_rbf = SVC(kernel='rbf', C=C_opt, gamma=gamma_opt).fit(X_tr,y_tr)
print('Accuracy score on validation set: ', svc_rbf.score(X_val, y_val))
```
